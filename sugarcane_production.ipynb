CORRELATION


```python

df_numeric = df[["Production (Tons)", "Production per Person (Kg)", "Acreage (Hectare)", "Yield (Kg / Hectare)"]].copy()
df_numeric["Production (Tons)"] = pd.to_numeric(df_numeric["Production (Tons)"], errors="coerce")
df_numeric["Production per Person (Kg)"] = pd.to_numeric(df_numeric["Production per Person (Kg)"], errors="coerce")
df_numeric["Acreage (Hectare)"] = pd.to_numeric(df_numeric["Acreage (Hectare)"], errors="coerce")
df_numeric["Yield (Kg / Hectare)"] = pd.to_numeric(df_numeric["Yield (Kg / Hectare)"], errors="coerce")
correlation_matrix = df_numeric.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Sugarcane Production Data')
plt.show()
```

```python
df_continent=df.groupby("Continent").sum()
```

```python
df_continent
```

```python
df_continent["Production (Tons)"] = pd.to_numeric(df_continent["Production (Tons)"], errors='coerce')
```

```python
df_continent["Production (Tons)"].sort_values(ascending=False).plot(kind="bar")
```

```python
df_continent["Acreage (Hectare)"] = pd.to_numeric(df_continent["Acreage (Hectare)"], errors='coerce')
```

```python
df_continent["Acreage (Hectare)"].sort_values(ascending=False).plot(kind="bar")
```

```python
df_continent["numbers of countries"]=df.groupby("Continent").count()["Country"]
```

```python
df_continent
```

```python
continent_names=df_continent.index.to_list()
sns.lineplot(data=df_continent,x="numbers of countries",y="Production (Tons)")
plt.xticks(df_continent["numbers of countries"],continent_names,rotation=90)
plt.show()
```
